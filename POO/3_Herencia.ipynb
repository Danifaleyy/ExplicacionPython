# ***Herencia***

 ***Que es la herencia?*** *Es heredar o pasarle todos los atributos y metodos de la clase padre a sus clases hijas.*

 ***Clase superclase o padre:*** *Da sus atributos y elementos a la subclase o clase hija.*

 ***Clase hija o subclase:*** *Recibe los atributos de la superclase o clase padre.*

 ***Importante:*** *Las subclases van a tener todos los atributos y mtodos de la superclase (Atributos y mtodos bsicos), pero se va a poder definir nuevos atributos y mtodos para cada subclase en especifica si lo deseas.*

 ***Para que sirve la herencia?*** *Para reutilizar cdigo de la clase padre (Atributos y elementos bsicos de la clase padre o Superclase) y ya solo agregas o defines los atributos y elementos adicionales a las nuevas clases (Clase hija o subclases).*

 ***Tipos de herencia:***

 ***Herencia simple:*** *Es la capacidad de una clase de heredar propiedades (atributos y mtodos) de una nica clase padre.*

 ***Herencia multiple:*** *Es la capacidad de una clase de heredar propiedades de ms de una clase padre.*





> ###  ***Herencia simple***

 ***Herencia simple:*** *Es la capacidad de una clase de heredar propiedades (atributos y mtodos) de una nica clase padre.*

 ***Estructura:***

        class ClaseHija(ClasePadre):
            pass

 ***Funcion super().__init__(atributo,atributo):*** *Se utiliza para agregar atributos y metodos extras a una clase hija*



```python
#Creando la clase padre
class Persona:
    #Atributos propios
    #Metodo __init__(self):
    #Como crear atributos: despues de poner el paramtro self, le pasas los nombres de los atributos que quieres que tenga esa clase
    #def __init__(self, atributo atributo, atributo):
    def __init__(self, nombre, edad, nacionalidad):
        #Definir atributos
        #self.atributo = atributo
        self.nombre = nombre
        self.edad = edad
        self.nacionalidad = nacionalidad
    
    #Como crear metodos: 
    #Siempre debes de poner el parametro self
    def hablar(self):
        print("Hola")

#Creando la clase hija:
#class ClaseHija(ClasePadre):
#Solamente se pone el nombre de la clase padre dentro del parentesis, asi te das cuenta de que es una clase hija
#Cuando no le pones contenido a la clase hija, solamente heredara los metodos y atribtuos de la clase padre
class Empleado(Persona):
    pass

#Creando un objeto de la clase hija: Le tienes que pasar la misma cantidad de valores al objeto que la misma cantidad de atributos que tiene el objeto, si no te dara error.
#nombre_objeto = NombreClase(valor_atributo, valor_atributo, valor_atributo...)
empleado1 = Empleado("Daniel", 21, "Mexicana")

#Accediendo a los atributos del objeto
print(empleado1.nombre)
print(empleado1.edad)
print(empleado1.nacionalidad)

#Accediendo al un metodo: Solo podemos acceder al metodo de un objeto, al metodo de una clase no se puede acceder
#print(nombre_objeto.nombre_metodo())
print(empleado1.hablar())
```

```python
#Creando la clase padre
class Persona:
    #Atributos propios
    #Metodo __init__(self):
    #Como crear atributos: despues de poner el paramtro self, le pasas los nombres de los atributos que quieres que tenga esa clase
    #def __init__(self, atributo atributo, atributo):
    def __init__(self, nombre, edad, nacionalidad):
        #Definir atributos
        #self.atributo = atributo
        self.nombre = nombre
        self.edad = edad
        self.nacionalidad = nacionalidad
    
    #Como crear metodos: 
    #Siempre debes de poner el parametro self
    def hablar(self):
        print("Hola")

#Creando la clase hija con herencia simple:
#Estructura: class ClaseHija(ClasePadre):
#Solamente se pone el nombre de la clase padre dentro del parentesis, asi te das cuenta de que es una clase hija
#Cuando no le pones contenido a la clase hija, solamente heredara los metodos y atribtuos de la clase padre
#Como agregar atributos y metodos extras a una clase hija: Se utiliza la funcion super().__init__(atributo,atributo)
class Empleado(Persona):
    #Atributos que quieres que tenga esta clase (Los que hereda y los que le quieres agregar a la clase hija)
    def __init__(self, nombre, edad, nacionalidad, puesto, salario):
        #Atributos que quiere que herede la clase hija de la clase padre
        super().__init__(nombre, edad, nacionalidad)
        #Se crean los atributos especificos de la clase hija
        self.puesto = puesto
        self.salario = salario

#Creando un objeto de la clase hija: Le tienes que pasar la misma cantidad de valores al objeto que la misma cantidad de atributos que tiene el objeto, si no te dara error.
#nombre_objeto = NombreClase(valor_atributo, valor_atributo, valor_atributo...)
empleado1 = Empleado("Daniel", 21, "Mexicana", "CEO", 100000)

#Accediendo a los atributos del objeto
print(empleado1.nombre)
print(empleado1.edad)
print(empleado1.nacionalidad)
print(empleado1.puesto)
print(empleado1.salario)
```

> ###  ***Herencia multiple***

 ***Herencia multiple:*** *Es la capacidad de una clase de heredar propiedades de ms de una clase padre.*

 ***IMPORTANTE:*** *Ya no se usa la funcion super porque es herencia multiple, se pone el nombre de la clase padre y los atributos de la clase padre que heredara.*

 ***Estructura:*** *ClasePadre.__init__(self, atributoclasepadre, atributoclasepadre...)*

 ***super vs self:*** 

 ***super:*** *Se utiliza para llamar al atributo de la clase padre*

 ***self:*** *Se utiliza para llamar al atributo de la clase actual*



```python
#Clase padre
class Persona:
    #Atributos propios
    #Metodo __init__(self):
    #Como crear atributos: despues de poner el paramtro self, le pasas los nombres de los atributos que quieres que tenga esa clase
    #def __init__(self, atributo atributo, atributo):
    def __init__(self, nombre, edad, nacionalidad):
        #Definir atributos
        #self.atributo = atributo
        self.nombre = nombre
        self.edad = edad
        self.nacionalidad = nacionalidad
    
    #Como crear metodos: 
    #Siempre debes de poner el parametro self
    def hablar(self):
        print("Hola, estoy hablando un poco")

#Clase padre
class Artista:
    #Atributos propios
    #Metodo __init__(self):
    #Como crear atributos: despues de poner el paramtro self, le pasas los nombres de los atributos que quieres que tenga esa clase
    #def __init__(self, atributo atributo, atributo):
    def __init__(self, habilidad):
        #Definir atributos
        #self.atributo = atributo
        self.habilidad = habilidad
    
    #Como crear metodos: 
    #Siempre debes de poner el parametro self
    def mostrar_habilidad(self):
        print(f"mi habilidad es: {self.habilidad}")

#Clase hija con herenci multiple

#Estructura: class ClaseHija(ClasePadre, ClasePadre):
#Solamente se pone el nombre de la clase padre dentro del parentesis, asi te das cuenta de que es una clase hija
#Cuando no le pones contenido a la clase hija, solamente heredara los metodos y atribtuos de la clase padre
#Como agregar atributos y metodos extras a una clase hija: Se utiliza la funcion super().__init__(atributo,atributo)
class EmpleadoArtista(Persona, Artista):
    #Atributos que quieres que tenga esta clase (Los que hereda y los que le quieres agregar a la clase hija)
    def __init__(self, nombre, edad, nacionalidad, habilidad, salario, empresa):
        #Atributos que quiere que herede la clase hija de la clase padre
        #Ya no se usa la funcion super porque es herencia multiple, se pone el nombre de la clase padre y los atributos de la clase padre que heredara.
        #Estructura: ClasePadre.__init__(self, atributoclasepadre, atributoclasepadre...)
        Persona.__init__(self, nombre, edad, nacionalidad)
        Artista.__init__(self, habilidad)
        #Se crean los atributos especificos de la clase hija
        self.salario = salario
        self.empresa = empresa
    
    #Como crear metodos: 
    #Siempre debes de poner el parametro self
    def presentarse(self):
        print(f"Hola, soy {self.nombre}, mi habilidad es {self.habilidad} y trabajo en la empresa {self.empresa}")

#Creando un objeto de la clase hija: Le tienes que pasar la misma cantidad de valores al objeto que la misma cantidad de atributos que tiene el objeto, si no te dara error.
#nombre_objeto = NombreClase(valor_atributo, valor_atributo, valor_atributo...)
roberto = EmpleadoArtista("Roberto", 43, "Argentino", "Cantar", 10000, "Google")

#Accediendo al metodo del objeto: objeto.metodo()
roberto.presentarse()

#Como saber si una clase es clase hija de otra clase:
#Estructura: variable_nueva = issubclass(ClaseHija, ClasePadre)
herencia = issubclass(EmpleadoArtista, Artista)
print(herencia)

#Como saber si un objeto es una instancia (Que se creo apartir de esa clase) de una clase
#Estructura: variable_nueva = isinstance(objeto, Clase)
instancia = isinstance(roberto, EmpleadoArtista)
print(instancia)
```
