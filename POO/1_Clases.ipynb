# ***Clase y objetos***

❓ ***Que es Programación orientada a objetos?*** *Es un estilo de programación.*

✅ ***Ventaja de usar POO:*** *La Programación Orientada a objetos permite que el código sea reutilizable, organizado y fácil de mantener.*

❓ ***¿Que es una clase?*** *Es un molde para crear objetos (De una clase puedes sacar muchos objetos: Porque una clase es tu molde genérico, tiene el mismo tipo de atributos y métodos pero con diferente valores o datos) (Diferentes acciones y características u propiedades).*

❓ ***¿Que es una objeto u instancia?*** *Un objeto u instancia tiene 2 propiedades:*

        ❓ 1- ¿Que es un atributo? Es una caracteristica del objeto

        ❓ 2- ¿Que es un metodo? Es una accion que puede realizar el objeto

💡 ***Todo en python es un objeto***

❓ ***Que es instanciación?*** *Un objeto es creado apartir de una clase, el proceso de la creacion del objeto se llama instanciación*

☢️ ***IMPORTANCIA DE LOS OBJETOS:*** *Los atributos y metodos de una clase no las podemos mostrar, para eso se crean los objetos (Solamente se puede mostrar cuando llamas a un objeto de esa clase), solamente existen en los objetos*

❗ ***Cuando modificas el atributo de un objeto solamente estas afectando a ese objeto y no a los otros objetos***

❗ ***A los atributos y metodos de una clase no se modifican o se le dan valores, para eso se crean los objetos, a los objetos si los podemos modifciarm una clase tiene valores vacios.***

❗ ***Los atributos y metodos creados para una clase estaran en todos los objetos creados.***

> ### 🟢 ***Como crear una clase***

✅ ***Regla para el nombre de una clase:*** *Cada palabra inicia con mayuscula (Ejemplo: CuentaBancaria)*

💡 *Usamos la palabra clave:* ***class***

💡 ***Estructura:***

        *class NombreClase:*

                *atributos*

                *metodos*

```python
class Celular:
    #Atributos fijos
    Nombre = "Iphone"
    Modelo = "14 pro max"
    Camara = "24mpx"
```

> ### 🟢 ***Como crear un objeto***

💡 ***Estructura:*** *nombre_objeto = NombreClase(Parametro1,Parametro2)*

💡***Como acceder a un atributo de un objeto:*** *print(nombre_objeto.nombre_atributo)*



```python
class Celular:
    #Atributos fijos
    Nombre = "Iphone"
    Modelo = "14 pro max"
    Camara = "24mpx"

#Creando objetos
celular1 = Celular()
celular2 = Celular()
celular3 = Celular()
#Mal
print(celular1)
print(celular2)
print(celular3)
#Como acceder a un atributo de un objeto: print(nombre_objeto.nombre_atributo)
print(celular1.Nombre)
print(celular2.Modelo)
print(celular3.Camara)
```

> ### 🟢 ***Atributos***

❓ ***¿Que es un atributo?*** *Es una caracteristica del objeto*

❓ ***Tipos de atributos:***

🟡 ***Atributos fijos:*** *Son atributos que se definen en la clase, tendran ese valor por default para todos los objetos que creen de la clase, no es reomendable*

🟡 ***Atributos propios:*** *Nos referimos a un atributo que ha sido definido dentro de la clase y que está disponible para cada instancia individual de esa clase. Estos atributos son específicos de cada instancia y pueden tener diferentes valores para cada objeto creado a partir de la misma clase.*

> ✅ ***Como crear atributos propios***

💡 ***Metodo __init__(self):*** *Sirve para que puedas crear objetos de esa clase, Este método se llama automáticamente cuando se crea un nuevo objeto a partir de una clase y se utiliza para realizar tareas de inicialización, como establecer valores predeterminados para los atributos de la instancia.*

💡 ***Parametro (self):*** *Hace referencia a la instancia actual del objeto en el que se está trabajando (Es una referencia al objeto mismo y permite acceder y manipular los atributos y métodos de esa instancia.)*

❗ ***A los atributos y metodos de una clase no se modifican o se le dan valores, para eso se crean los objetos, a los objetos si los podemos modifciarm una clase tiene valores vacios.***

❗ ***Los atributos y metodos creados para una clase estaran en todos los objetos creados.***

> ☢️ ***IMPORTANCIA DE LOS OBJETOS:*** *Los atributos y metodos de una clase no las podemos mostrar, para eso se crean los objetos (Solamente se puede mostrar cuando llamas a un objeto de esa clase), solamente existen en los objetos*

✅ ***Llamando un atributo de un objeto***

```python
class Celular:
    #Atributos fijos
    Nombre = "Iphone"
    Modelo = "14 pro max"
    Camara = "24mpx"

#Creando objetos
celular1 = Celular()
celular2 = Celular()
celular3 = Celular()

#Como acceder a un atributo de un objeto: print(nombre_objeto.nombre_atributo)
print(celular2.Modelo)
```

> ### 🟢 ***Metodos***

❓ ***¿Que es un metodo?*** *Es una accion que puede realizar el objeto*

❓ ***Metodos vs Funciones:*** *Un metodo es una funcion, si la funcion se crea dentro de una clase se le llamara metodo y si se crea fuera de la clase se le llamara funcion.*

> ☢️ ***Como crear metodos***

💡 ***Estructura:*** *Siempre debes de poner el parametro self*

        class Celular:

                def nombre_metodo(self):
            
                    Codigo

❗ ***A los atributos y metodos de una clase no se modifican o se le dan valores, para eso se crean los objetos, a los objetos si los podemos modifciarm una clase tiene valores vacios.***

❗ ***Los atributos y metodos creados para una clase estaran en todos los objetos creados.***

> ☢️ ***Crear objetos con atributos propios***

```python
#Creando una clase
class Celular:
    #Atributos propios
    #Metodo __init__(self):
    #Como crear atributos: despues de poner el paramtro self, le pasas los nombres de los atributos que quieres que tenga esa clase
    #def __init__(self, atributo atributo, atributo):
    def __init__(self, marca, modelo, camara):
        #Definir atributos
        #self.atributo = atributo
        self.marca = marca
        self.modelo = modelo
        self.camara = camara
        
    #Como crear metodos: 
    #Siempre debes de poner el parametro self
    def llamar(self):
        print("Estas llamando")
    def cortar(self):
        print("cortaste la llamada")
         
#Creando un objeto de la clase: Le tienes que pasar la misma cantidad de valores al objeto que la misma cantidad de atributos que tiene el objeto, si no te dara error.
#nombre_objeto = NombreClase(valor_atributo, valor_atributo, valor_atributo...)
celular1 = Celular("Samsung", "S23", "128mpx")
print(celular1.marca)

#Creando otro objeto
celular2 = Celular("Huawei", "Mate 60", "256mpx")
print(celular2.marca)

#Accediendo al un metodo: Solo podemos acceder al metodo de un objeto, al metodo de una clase no se puede acceder
#print(nombre_objeto.nombre_metodo())
print(celular1.llamar())
print(celular1.cortar())
```

```python
#Creando una clase
class Celular:
    #Atributos propios
    #Metodo __init__(self):
    #Como crear atributos: despues de poner el paramtro self, le pasas los nombres de los atributos que quieres que tenga esa clase
    #def __init__(self, atributo atributo, atributo):
    def __init__(self, marca, modelo, camara):
        #Definir atributos
        #self.atributo = atributo
        self.marca = marca
        self.modelo = modelo
        self.camara = camara
        
    #Como crear metodos: 
    #Siempre debes de poner el parametro self
    #Usando f string:
    def llamar(self):
        print(f'Estas llamando desde un: {self.modelo}')
    def cortar(self):
        print(f'Cortaste la llamada desde un: {self.modelo}')
         
#Creando un objeto de la clase: Le tienes que pasar la misma cantidad de valores al objeto que la misma cantidad de atributos que tiene el objeto, si no te dara error.
#nombre_objeto = NombreClase(valor_atributo, valor_atributo, valor_atributo...)
celular1 = Celular("Samsung", "S23", "128mpx")
print(celular1.marca)

#Creando otro objeto
celular2 = Celular("Huawei", "Mate 60", "256mpx")
print(celular2.marca)

#Accediendo al un metodo: Solo podemos acceder al metodo de un objeto, al metodo de una clase no se puede acceder
#print(nombre_objeto.nombre_metodo())
print(celular1.llamar())
print(celular1.cortar())
```
