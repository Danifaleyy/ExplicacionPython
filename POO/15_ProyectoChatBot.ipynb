```python
#Esta biblioteca es de procesamiento de lenguaje pero en ingles
from textblob import TextBlob

class AnalizadorDeSentimientos:
    def analizar_sentimientos(self, texto):
        analisis = TextBlob(texto)
        #.sentimient: Es una propiedad del objeto analisis
        #.polarity: Es una propiedad de la propeidad sentimient
        if analisis.sentiment.polarity > 0:
            return "Positivo"
        elif analisis.sentiment.polarity == 0:
            return "Neutral"
        else:
            return "Negativo"

analizador = AnalizadorDeSentimientos()
#El texto que le ingreses debe de ser en ingles para que funcione
resultado = analizador.analizar_sentimientos("I am good")
print(resultado)
```
