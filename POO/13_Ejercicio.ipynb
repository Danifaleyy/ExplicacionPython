```python
class Personaje:
    def __init__(self, nombre, fuerza, velocidad):
        self.nombre = nombre
        self.fuerza = fuerza
        self.velocidad = velocidad
    
    def __repr__(self):
        return f"Nombre: {self.nombre}, Fuerza: {self.fuerza}, Velocidad: {self.velocidad}"
    
    #Que pasa cuando lo sumamos: Sobrecarga de operadores
    def __add__(self, otro_personaje):
        nuevo_nombre = self.nombre + " " + otro_personaje.nombre
        nueva_fuerza = ((self.fuerza + otro_personaje.fuerza)/2)**2
        nueva_velocidad = ((self.velocidad + otro_personaje.velocidad)/2)**2
        return Personaje(nuevo_nombre, nueva_fuerza, nueva_velocidad)
        
goku = Personaje("Goku", 999, 10)
print(goku)
vegeta = Personaje("Vegeta", 1000, 10)
print(vegeta)
gogeta = goku + vegeta
print(gogeta)
suma_fuerza = goku.fuerza + vegeta.fuerza
print(f"Suma de la fuerza: {suma_fuerza}")
```
